```python
import random
from random import randint
from math import log
import numpy as np
import datetime
import scipy.io
time0 = datetime.datetime.now()
```

```python
def norm(vect):
    sqr_sum = np.sum(np.vectorize(lambda x: np.abs(x)**2)(vect))**(1/2)
    vect = np.vectorize(lambda x: x/sqr_sum)(vect)
    return vect
@njit
def conj(vect):
    return np.conj(vect)
@njit
def shift5(arr, num, fill_value=np.nan):
    result = np.empty_like(arr)
    if num > 0:
        result[:num] = fill_value
        result[num:] = arr[:-num]
    elif num < 0:
        result[num:] = fill_value
        result[:num] = arr[-num:]
    else:
        result = arr
    return result
@njit
def shift(AR, step):
    if step >= 0:
        return shift5(AR, step, 0)+ shift5(AR, step-len(AR), 0)
    else:
        return shift5(AR, step, 0)+ shift5(AR, len(AR)+step, 0)
```
